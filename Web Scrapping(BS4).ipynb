# WEB SCRAPING using Beautiful Soup(BS4)

Importing Libraries

```python
import bs4
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

For this task, we will be scraping Mobile Data from Flipkart Website. We will be storing the link of the product in the variable Url.

```python
Url  = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_3_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_3_0_na_na_na&as-pos=3&as-type=TRENDING&suggestionId=mobiles&requestId=852dff55-7c12-41f0-81e6-e9a09f4cf0ce'
```

 Send HTTP request to the URL of the page you want to scrape

```python
page = requests.get(Url)
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find('div',class_="_4rR01T")
print(name)
```

```python
#get rating of a product
rating=soup.find('div',class_="_3LWZlK")
print(rating)
rating.text
```

```python
#get other details and specifications of the product
specification=soup.find('div',class_="fMghEO")
print(specification)
specification.text
```

```python
for each in specification:
    spec=each.find_all('li',class_='rgWa7D')
    print(spec[0].text)
    print(spec[1].text)
    print(spec[2].text)
    print(spec[4].text)
```

```python
#get price of the product
price=soup.find('div',class_='_30jeq3 _1_WHN1')
print(price)
price.text
```

```python
products=[]              
prices=[]                
ratings=[]               
Display = []                             
memory = []              
processor = []           
camera = []              
```

```python
for data in soup.findAll('div',class_='_3pLy-c row'):
        names=data.find('div', attrs={'class':'_4rR01T'})
        price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})
        specification = data.find('div', attrs={'class':'fMghEO'})
        
        for each in specification:
            col=each.find_all('li', attrs={'class':'rgWa7D'})
            Memory_ =col[0].text
            Display_ = col[1].text
            camera_ = col[2].text
            processor_ = col[3].text
            
        products.append(names.text) # Add product name to list
        prices.append(price.text) # Add price to list
        memory.append(Memory_)# Add supported apps specifications to list
        Display.append(Display_) # Add operating system specifications to list
        camera.append(camera_) # Add resolution specifications to list
        processor.append(processor_) # Add sound specifications to list
        ratings.append(rating.text)   #Add rating specifications to list
```

```python
import pandas as pd
df=pd.DataFrame({'Product Name':products,'Price':prices,'Memory':memory,'Display':Display,"CAMERA":camera,'Rating':ratings})
df.head(10)
```
